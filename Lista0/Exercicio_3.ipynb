```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
x = np.random.randn(1000,1)
w = np.random.normal(0, 1, (1000,1))
```

```python
y = 1.2 + 2.3*x + 10*w
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, '.')
ax.set(xlabel='x', ylabel='y')
ax.grid()
```
