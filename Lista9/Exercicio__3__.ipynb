```python
# Import all necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=5, cluster_std=1.0, shuffle=True,
random_state=42)
```

```python
y = y % 2
plt.figure()
plt.title('Dataset para clustering')
plt.scatter(X[:,0], X[:,1],marker= 'o', s=30, color = 'black')
plt.show()
```

Letra B) Observando a figura, podemos notar que devem ser usados um total de 4 clusters para separar os dados de treinamento.

```python
init_clusters = 'k-means++'
# Fit K-means with Scikit
km = KMeans(n_clusters=4, init=init_clusters)
km.fit(X)
```

```python
# Predict the cluster for all the samples
P = km.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=P, s=50, cmap='viridis')

centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
