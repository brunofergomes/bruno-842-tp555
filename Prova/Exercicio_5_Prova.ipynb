Cálculo da entropia para o conjunto total de treinamento:

Contando na tabela o número total de dados, com relação às classes mencionadas, temos que:

Classe Pagou: 12/17
Classe Não Pagou: (1 - 12/17) = 5/17

Portanto, é calculada a entropia do conjunto total de treinamentos como:

H(12+, 5-) = -( (12/17)* log2(12/17) + (5/17)* log2(5/17) )

H(12+, 5-) = 0.874

Em posse da entropia referente às classes, podemos calcular agora os ganhos de informação referentes a cada atributo. Porém, para realizar este cálculo, devemos reduzir as tabelas, e refêrencia-las ao atributo específico.

Com isso, seguimos calculando os ganhos:

Método: G(x) = H(12+,5-) - sum( (Sae/St)* H(Spe/Sae) )

Sae = S atributo especifico; St = S total; Spe = S positivo especifico

G(PC) = 0.874 - ( (7/17)* H(6/7) + (10/17)* H(6/10) ) = 0.0592

G(EC) = 0.874 - ( (5/17)* H(4/5) + (8/17)* H(6/8) + (4/17)* H(2/4) ) = 0.0446

G(Xp) = 0.874 - ( (0/17)* H(0/0) + (0/17)* H(0/0) + (6/17)* H(3/6) + (6/17)* H(4/6) + (4/17)* H(4/4) + (1/17)* H(1/1) ) = 0.197

Podemos notar então que o ganho de informação é maximizado com o atributo Xp, que é então o nó raiz da árvore!
