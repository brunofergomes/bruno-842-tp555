```python
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
X = tf.placeholder(tf.float32, shape=[4,2], name = 'X')
Y = tf.placeholder(tf.float32, shape=[4,1], name = 'Y')

W = tf.Variable(tf.truncated_normal([2,2]), name = "W")
w = tf.Variable(tf.truncated_normal([2,1]), name = "w")

c = tf.Variable(tf.zeros([4,2]), name = "c")
b = tf.Variable(tf.zeros([4,1]), name = "b")

# XOR data
x = [[0,0],[0,1],[1,0],[1,1]]
y = [[0],[1],[1],[0]]
```

# Caso 1 : Gradient Descendent Optimizer

```python
with tf.name_scope("hidden_layer") as scope:
    h = tf.nn.relu(tf.add(tf.matmul(X, W),c))

with tf.name_scope("output") as scope:
    y_estimated = tf.sigmoid(tf.add(tf.matmul(h,w),b))

with tf.name_scope("loss") as scope:
    loss = tf.reduce_mean(tf.squared_difference(y_estimated, Y)) 

with tf.name_scope("train") as scope:
    train_step = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()

writer = tf.summary.FileWriter("./logs/xor_logs", sess.graph)

sess.run(init)
```

```python
for epoch in range(40001):
    sess.run(train_step, feed_dict={X: x, Y: y})
    if epoch % 100 == 0:
        print("_"*80)
        print('Epoch: ', epoch)
        print('   loss: ', sess.run(loss, feed_dict={X: x, Y: y}))
```

```python
# Reset graph for new Optimizer
reset_graph()
```

# Caso 2 : Adam Optimizer

```python
X = tf.placeholder(tf.float32, shape=[4,2], name = 'X')
Y = tf.placeholder(tf.float32, shape=[4,1], name = 'Y')

W = tf.Variable(tf.truncated_normal([2,2]), name = "W")
w = tf.Variable(tf.truncated_normal([2,1]), name = "w")

c = tf.Variable(tf.zeros([4,2]), name = "c")
b = tf.Variable(tf.zeros([4,1]), name = "b")
```

```python
with tf.name_scope("hidden_layer") as scope:
    h = tf.nn.relu(tf.add(tf.matmul(X, W),c))

with tf.name_scope("output") as scope:
    y_estimated = tf.sigmoid(tf.add(tf.matmul(h,w),b))

with tf.name_scope("loss") as scope:
    loss = tf.reduce_mean(tf.squared_difference(y_estimated, Y)) 

with tf.name_scope("train") as scope:
    train_step = tf.train.AdamOptimizer(learning_rate=0.01).minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()

writer = tf.summary.FileWriter("./logs/xor_logs", sess.graph)

sess.run(init)
```

```python
for epoch in range(40001):
    sess.run(train_step, feed_dict={X: x, Y: y})
    if epoch % 100 == 0:
        print("_"*80)
        print('Epoch: ', epoch)
        print('   loss: ', sess.run(loss, feed_dict={X: x, Y: y}))
```

Observando a resposta de perdas referentes aos dois otimizadores implementados, podemos chegar a conclusão de que o Adam Optmizer obteve melhor performance quando comparado com o Gradient Descendent Optimizer. Considerando a última época após o treinamento (Epoch:40000), o Adam Optimizer conseguiu convergir com erro mínimo de aproximadamente 8.94e-11 enquanto que o Gradient Descendent Optimizer convergiu com erro mínimo de aproximadamente 0.0016 .

Melhor Performance: Adam Optimizer
